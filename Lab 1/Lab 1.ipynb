### Завдання 1. 
Зобразити поле напрямків у точках $(m, \frac{n}{2})$, де $m = {-6 , -5, \ldots , 6}, \enspace  n = {-6 , -5, \ldots , 6}$

```python
def diff(x,y):
    return y*y - x

plt.rcParams["figure.figsize"] = (20,10)

m = np.linspace(-6,6,13)
n = np.linspace(-6,6,13)/2

for j in m:
    for k in n:
        slope = diff(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')

plt.title("Slope field y'")
plt.grid(True)
plt.show()
```
